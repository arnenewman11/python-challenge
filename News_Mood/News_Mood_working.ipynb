```python
#loop through search terms
target_terms = ["@BBC", "@CBS", "@CNN", "@Fox", "@NYT"]

# Lists to hold sentiments
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
source_list = []
tweets_text = []
tweet_dates = []

for term in target_terms:

    # Grab 25 tweets
    public_tweets = api.search(term, count=100, result_type="recent", lang = "en")

    # Loop through all tweets
    for tweet in public_tweets["statuses"]:

        # Run Vader Analysis on each tweet
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        source = str.replace(term,"@","")
        tweet_text = tweet["text"]
        tweet_date = tweet["created_at"]

        # Add each value to the appropriate array
        compound_list.append(compound)
        positive_list.append(pos)
        negative_list.append(neg)
        neutral_list.append(neu)
        source_list.append(source)
        tweets_text.append(tweet_text)
        tweet_dates.append(tweet_date)

# Store the Average Sentiments
sentiment = {"Compound": compound_list,
             "Positive": positive_list,
             "Neutral": negative_list,
             "Negative": neutral_list,
             "Source":source_list,
             "Tweet Text": tweets_text,
             "Tweet Date": tweet_dates}

sentiment_df = pd.DataFrame(sentiment)
    
sentiment_df
```
